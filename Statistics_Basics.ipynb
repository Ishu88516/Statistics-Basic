#Statistics Basics

Question 1

What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer:

Statistics is broadly classified into descriptive statistics and inferential statistics.

Descriptive statistics deals with methods used to summarize, organize, and present data in a meaningful way. It does not draw conclusions beyond the given data. Common measures include mean, median, mode, charts, and graphs.
Example: Calculating the average marks of students in a class.

Inferential statistics involves methods used to draw conclusions or make predictions about a population based on a sample. It includes hypothesis testing, confidence intervals, correlation, and regression.
Example: Using a sample survey to estimate customer satisfaction of all customers.

Question 2

What is sampling in statistics? Explain the differences between random sampling and stratified sampling.

Answer:

Sampling is the process of selecting a subset of individuals or observations from a population to represent the entire population.

Random sampling is a method where every member of the population has an equal chance of being selected. It is simple and reduces bias.
Example: Randomly selecting 50 students from a college.

Stratified sampling divides the population into homogeneous subgroups (strata) and samples are taken from each group. This ensures better representation.
Example: Selecting students separately from each department.

Question 3

Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer:

Mean is the average of all values, calculated by dividing the sum of values by the number of values.

Median is the middle value when the data is arranged in ascending or descending order.

Mode is the value that occurs most frequently in a dataset.

These measures are important because they summarize large data into a single representative value, making comparison and interpretation easier. Mean shows overall performance, median avoids the effect of extreme values, and mode identifies the most common value.

Question 4

Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer:

Skewness measures the asymmetry of a data distribution.

Positive skew → tail on the right side

Negative skew → tail on the left side

A positive skew implies that most values are concentrated on the lower side, with a few extremely high values.

Kurtosis measures the peakedness or flatness of a distribution. High kurtosis indicates a sharp peak, while low kurtosis indicates a flat distribution.

Question 5

Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

Answer (Python Code & Output):
import statistics as stats

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

print("Mean:", stats.mean(numbers))
print("Median:", stats.median(numbers))
print("Mode:", stats.mode(numbers))


Output:

Mean: 19.6
Median: 19
Mode: 12

Question 6

Compute the covariance and correlation coefficient between the following two datasets:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

Answer (Python Code & Output):
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

print("Covariance:", np.cov(list_x, list_y)[0][1])
print("Correlation:", np.corrcoef(list_x, list_y)[0][1])


Output:

Covariance: 275.0
Correlation: 0.9959

Question 8

You are working as a data analyst in an e-commerce company. Explain how covariance and correlation are used to study advertising spend and daily sales. Also compute correlation using Python.

Answer:

Covariance helps determine the direction of the relationship, while correlation measures the strength and direction of the relationship between advertising spend and sales.

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

print("Correlation:", np.corrcoef(advertising_spend, daily_sales)[0][1])


Output:

Correlation: 0.9936


This indicates a strong positive relationship between advertising spend and sales.

Question 9

Customer satisfaction survey data is collected on a scale of 1–10. Explain suitable summary statistics and visualizations. Draw a histogram using Python.

Answer:

Summary statistics such as mean, median, standard deviation, and mode are used to understand central tendency and variability. A histogram helps visualize the distribution.

import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores)
plt.xlabel("Scores")
plt.ylabel("Frequency")

The histogram shows most values between 7 and 9, indicating high customer satisfaction.
